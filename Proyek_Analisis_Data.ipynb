### Cleaning Data

```python
# Menghapus kolom yang tidak diperlukan
cleaned_aotizhongxin_df = aotizhongxin_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_changping_df = changping_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_dingling_df = dingling_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_dongsi_df = dongsi_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_guanyuan_df = guanyuan_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_gucheng_df = gucheng_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_huairou_df = huairou_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_nongzhanguan_df = nongzhanguan_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_shunyi_df = shunyi_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_tiantan_df = tiantan_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_wanliu_df = wanliu_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
cleaned_wanshouxigong_df = wanshouxigong_df.drop(columns=["No", "TEMP", "PRES", "DEWP", "RAIN", "wd", "WSPM"])
```

```python
# Contoh hasil setelah menghapus kolom yang tidak diperlukan
cleaned_aotizhongxin_df.info()
```

```python
# Penanganan missing value
cleaned_aotizhongxin_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_changping_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_dingling_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_dongsi_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_guanyuan_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_gucheng_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_huairou_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_nongzhanguan_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_shunyi_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_tiantan_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_wanliu_df.interpolate(method="linear", limit_direction="both", inplace=True)
cleaned_wanshouxigong_df.interpolate(method="linear", limit_direction="both", inplace=True)
```

```python
cleaned_aotizhongxin_df.info()
```

```python
# Menggabungkan semua data
data_df = pd.concat([cleaned_aotizhongxin_df, cleaned_changping_df, cleaned_dingling_df, cleaned_dongsi_df, cleaned_guanyuan_df, cleaned_gucheng_df, cleaned_huairou_df, cleaned_nongzhanguan_df, cleaned_shunyi_df, cleaned_tiantan_df, cleaned_wanliu_df, cleaned_wanshouxigong_df], axis=0)
```

```python
data_df.info()
```

```python
data_df.to_csv("./data/cleaned_data.csv", index=False)
```

## Exploratory Data Analysis (EDA)

### Explore "Berdasarkan waktu dan jenis polutannya, kapan kualitas udara paling buruk?"

```python
# Mengambil data polutan
correlation_df = data_df.drop(columns=["year", "month", "day", "hour", "station"])
```

```python
# Menghitung matriks korelasi
correlation = correlation_df.corr()

# Menampilkan matriks korelasi
print(correlation)

# Visualisasi matriks korelasi dengan heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm", fmt='.2f')
plt.title("Correlation Matrix of Pollutants")
plt.show()
```

```python
air_quality_year = data_df.groupby("year").agg({"PM2.5": "mean", "PM10": "mean", "SO2": "mean", "NO2": "mean", "CO": "mean", "O3": "mean"})
air_quality_year
```

```python
air_quality_month = data_df.groupby("month").agg({"PM2.5": "mean", "PM10": "mean", "SO2": "mean", "NO2": "mean", "CO": "mean", "O3": "mean"})
air_quality_month.index = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
air_quality_month
```

### Explore "Stasiun mana yang memiliki kualitas udara terbaik dan terburuk?"

```python
air_quality_station = data_df.groupby("station").agg({"PM2.5": "mean", "PM10": "mean", "SO2": "mean", "NO2": "mean", "CO": "mean", "O3": "mean"})
air_quality_station
```

```python
# Menghitung rata-rata dan mengurutkan nilai mean
air_quality_station_mean = air_quality_station.drop(columns=["O3"]).mean(axis=1)
air_quality_station_mean_df = air_quality_station_mean.to_frame(name="Mean Concentration")
air_quality_station_mean_df
```

**Insight:**
- Kualitas Udara dipengaruhi oleh konstentrasi polutan. Semakin tinggi konsentrasi PM2.5, PM10, SO2, NO2, dan CO, semakin buruk kualitas udaranya. Semakin tinggi konsentrasi O3, semakin baik kualitas udaranya.
- Berdasarkan konsentrasi 
- Stasiun Wanshouxigong memiliki kualitas data yang terburuk berdasarkan rata-rata konsentrasi pulutan berbahaya

```python
# Visualisasi Kualitas Udara Berdasarkan Tahun dan Polutan
all_air_quality_year = air_quality_year.drop(columns=["O3"])
all_air_quality_year.index = all_air_quality_year.index.astype(str)
all_air_quality_year_mean = all_air_quality_year.mean(axis=1).sort_values(ascending=False)
all_air_quality_year_mean_df = all_air_quality_year_mean.to_frame(name="Mean Concentration")
plt.figure(figsize=(10, 6))
sns.barplot(x="Mean Concentration", y=all_air_quality_year_mean_df.index, data=all_air_quality_year_mean_df, color="skyblue")
plt.title("Worse Air Quality by Year (Polutant Mean Concentration)")
plt.xlabel("Polutant Concentration (g/m)")
plt.ylabel("Year")
plt.show()

```
